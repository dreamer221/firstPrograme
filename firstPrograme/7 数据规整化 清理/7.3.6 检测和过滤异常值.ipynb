# 7.3.6 检测和过滤异常值

```python
data = pd.DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
data.head()
```

```python
# 找第三列中，绝对值大于3的数字：
col = data[2]
col[np.abs(col) > 3]
```

```python
# 找到所有数据中大于3数据所在的行
print((np.abs(data) > 3).any()) # 默认是0, 表示index,任意行中含有大于3的列
(np.abs(data) > 3).any(1) # axis=1 columns表示任意一个column中含有大于3的行
```

```python
# 所有绝对值大于3的数据所在的行
data[(np.abs(data) > 3).any(1)]
```

下面是把绝对值大于3的数字直接变成-3或3：

```python
# np.abs(data) > 3  # 1000 rows × 4 columns
data[np.abs(data) > 3] # 1000 rows × 4 columns, 超出我的想象
```

```python
data[np.abs(data) > 3] = np.sign(data) * 3
data.describe()
```

np.sign(data)会根据值的正负号来得到1或-1：
