# 9.3.0 应用：通用的分割-应用-合并

```python
import numpy as np
np.random.seed(9)
import pandas as pd
```

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'], 
                   'data1': np.random.randn(5),
                   'data2': np.random.randn(5)})
df
```

```python
k1_means = df.groupby('key1').mean().add_prefix('mean_')
k1_means
```

```python
pd.merge(df, k1_means, left_on='key1', right_index=True)
```

```python
keys = ['one', 'two', 'one', 'two', 'one']
people = pd.DataFrame(np.random.randn(5, 5),
                      columns=['a', 'b', 'c', 'd', 'e'],
                      index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
people
```

```python
for key, table in people.groupby(keys):
    print(key)
    print(table, '\n')
```

```python
people.groupby(keys).mean()
```

```python
people.groupby(keys).transform(np.mean)
```

```python
def demean(arr):
    return arr - arr.mean()

demeand = people.groupby(keys).transform(demean)
demeand
```

```python
demeand.groupby(keys).mean() ## 都是零
```

# 9.3.1 一般性的 拆分-应用-合并

```python
tips = pd.read_csv('../examples/tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]

top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=3, column='total_bill')
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
print(type(result), result.index, result.columns, sep='\n')
result
```

在GroupBy内部，当我们想要调用一个像describe这样的函数的时候，其实相当于下面的写法：

    f = lambda x: x.describe()
    grouped.apply(f)

```python
result.unstack('smoker')
```

### 抑制组键

```python
tips.groupby('smoker', group_keys=False).apply(top)
```

```python
tips.groupby('smoker').apply(top)
```

# 9.3.2 分位数与桶分析

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),
                       'data2': np.random.randn(1000)})
frame.head()
```

```python
quartiles = pd.cut(frame.data1, 4)
quartiles[:10]
```

cut返回的Categorical object（类别对象）能直接传入groupby。所以我们可以在data2列上计算很多统计值：

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}
```

```python
grouped = frame.data2.groupby(quartiles)
```

```python
grouped.apply(get_stats).unstack()
```

也有相同长度的桶（equal-length buckets）；想要按照样本的分位数得到相同长度的桶，用qcut。这里设定labels=False来得到分位数的数量：

```python
# Return quantile numbers
grouping = pd.qcut(frame.data1, 10, labels=False)
```

译者：上面的代码是把frame的data1列分为10个bin，每个bin都有相同的数量。因为一共有1000个样本，所以每个bin里有100个样本。grouping保存的是每个样本的index以及其对应的bin的编号。

```python
grouped = frame.data2.groupby(grouping)
```

```python
grouped.apply(get_stats).unstack()
```
