```python
import numpy as np
np.random.seed(9)
import pandas as pd
```

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'], 
                   'data1': np.random.randn(5),
                   'data2': np.random.randn(5)})
df
```

```python
k1_means = df.groupby('key1').mean().add_prefix('mean_')
k1_means
```

```python
pd.merge(df, k1_means, left_on='key1', right_index=True)
```

```python
keys = ['one', 'two', 'one', 'two', 'one']
people = pd.DataFrame(np.random.randn(5, 5),
                      columns=['a', 'b', 'c', 'd', 'e'],
                      index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
people
```

```python
for key, table in people.groupby(keys):
    print(key)
    print(table, '\n')
```

```python
people.groupby(keys).mean()
```

```python
people.groupby(keys).transform(np.mean)
```

```python
def demean(arr):
    return arr - arr.mean()

demeand = people.groupby(keys).transform(demean)
demeand
```

```python
demeand.groupby(keys).mean() ## 都是零
```
