# 5.5.2 重排分级顺序

```python
df = pd.DataFrame(np.arange(12).reshape(4, 3),
                 index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                 columns=[['Ohio', 'Ohio', 'Colorado'], ['Green', 'Red', 'Green']])
df.index.names = ["key1", 'key2']
df.columns.names = ["state", 'color']
df
```

```python
df2 = df.swaplevel('key1', 'key2') # === df.swaplevel(0, 1)
df2
```

```python
df # 原始数据不变
```

```python
df.sort_index(level=1) # 默认axis=0=index, 按照key1排序
```

```python
df.swaplevel(0, 1).sort_index(level=1) # 默认axis=0=index，按照key2排序
```

```python
df
```

# 5.5.3 根据级别汇总

```python
df = pd.DataFrame(np.arange(12).reshape(4, 3),
                 index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                 columns=[['Ohio', 'Ohio', 'Colorado'], ['Green', 'Red', 'Green']])
df.index.names = ["key1", 'key2']
df.columns.names = ["state", 'color']
df
```

```python
df.sum(level=0)
```

```python
df.sum(level=1) # 
```

```python
df.sum(level='color', axis=1)
```

# 5.5.4 df行与列的互换

```python
df = pd.DataFrame({'a': range(7), 'b': range(7, 0, -1), \
                   'c': ['one'] * 3 + ['two'] * 4, 'd': [0, 1, 2, 0, 1, 2, 3]})
df
```

```python
df2 = df.set_index(['c', 'd']) # 创建新的df 将列转换为行索引，并删除
df2
```

```python
df3 = df2.reset_index()  # reset_index()将行索引转换为列,并删除行索引
df3
```

```python
df4 = df.set_index(['c', 'd'], drop=False)  # 不删除原始的列
df4
```
