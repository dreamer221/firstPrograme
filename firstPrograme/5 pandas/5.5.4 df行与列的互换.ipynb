# 5.5.4 df行与列的互换

```python
df = pd.DataFrame({'a': range(7), 'b': range(7, 0, -1), \
                   'c': ['one'] * 3 + ['two'] * 4, 'd': [0, 1, 2, 0, 1, 2, 3]})
df
```

```python
df2 = df.set_index(['c', 'd']) # 创建新的df 将列转换为行索引，并删除
df2
```

```python
df3 = df2.reset_index()  # reset_index()将行索引转换为列,并删除行索引
df3
```

```python
df4 = df.set_index(['c', 'd'], drop=False)  # 不删除原始的列
df4
```
